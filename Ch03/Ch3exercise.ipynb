Exercise 1:

```python
import pandas as pd

faang = pd.DataFrame()
for ticker in ['fb', 'aapl', 'amzn', 'nflx', 'goog']:
    df = pd.read_csv(f'faang/{ticker}.csv')
    # make the ticker the first column
    df.insert(0, 'ticker', ticker.upper())
    faang = faang.append(df)

```

Exercise 2:

```python
faang = faang.assign(
    date = lambda x: pd.to_datetime(x.date),
    volume=lambda x: x.volume.astype(int)
)
faang.sort_values(
    ['date', 'ticker']
)
faang.head()
```

Exercise 3:

```python
faang.nsmallest(7,'volume')
```

Exercise 4:

```python
melted_df= faang.melt(
    id_vars=['date', 'ticker'],
    value_vars=['open','high', 'low', 'close', 'volume']
)
melted_df.head()
```
